```python
import matplotlib.pyplot as plt


threads = [1,2, 4, 8]

time_test1 = [732.348, 416.443, 285.819, 239.889]
time_test2 = [698.445, 409.999, 262.634, 219.477]

speedup1 = [time_test1[0] / t for t in time_test1]
speedup2 = [time_test2[0] / t for t in time_test2]

plt.figure(figsize=(7, 5))
plt.plot(threads, speedup1, marker='o', label="Test 1")
plt.plot(threads, speedup2, marker='s', label="Test 2")
plt.plot(threads, threads, '--', label="Ideal")

plt.xlabel("Number of threads")
plt.ylabel("Speedup")
plt.title("Speedup vs number of threads")
plt.xticks(threads)
plt.legend()
plt.grid(True)

plt.show()
```
